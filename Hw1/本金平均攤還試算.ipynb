# 本金平均攤還試算  
1. 本金平均攤還法:  
 全部貸款本金每月平均分攤，利息則按貸款餘額逐期計算(每月償付之利息金額則逐月遞減)。  
 
2. 計算方式:  
 每月應攤付本金金額＝貸款本金÷還款總月數  
    每月應付利息金額＝本金餘額×月利率

## 使用方式
1. 在 <1>input 輸入本金、貸款期數、年利率 
2. 執行 <2><3><4> 即可印出試算結果

```python
# <1>
# input
F = 1200000  # 本金
year = 5     # 貸款期(年)數
rate = 0.05  # 年利率
```

```python
# <2>
import pandas as pd
```

```python
# <3>
# 計算每期應付本金
monthly_F = round(F / (year*12))

# for DataFrame
month_F = [monthly_F] * (year*12)
n = ['第'+str(i)+'期' for i in range(1, (year*12)+1)]
month_interest = []
acc_payment = []

# 計算每期應攤還利息 及 本息累計
accumulate_payment = 0
for i in range(year*12):
    interest = round((F-monthly_F*i)*(rate/12))
    accumulate_payment += monthly_F + interest
    month_interest.append(interest)
    acc_payment.append(accumulate_payment)

# 建立 DataFrame
dict = {"..": n,
        "本金(元)": month_F,
        "利息(元)": month_interest,
        "本息累計(元)": acc_payment}
output = pd.DataFrame(dict)
```
